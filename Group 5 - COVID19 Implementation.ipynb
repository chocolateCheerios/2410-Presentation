# Part Two

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Covid.csv')
df.drop(columns=['Data As Of', 'Start Date', 'End Date', 'Group', 'Year', 'Month', 'Age Group', 'ICD10_codes', 'Number of Mentions', 'Flag'], inplace=True)

# Numerical values
state_counts = df.groupby('State')['COVID-19 Deaths'].sum().reset_index()
# Data Cleaning
state_counts = state_counts[state_counts['State'] != 'United States']
state_counts = state_counts[state_counts['State'] != 'New York City']
state_counts = state_counts[state_counts['State'] != 'New York']
state_counts = state_counts.sort_values('COVID-19 Deaths', ascending=False)

fig, ax = plt.subplots(figsize=(20, 10))
state_counts.plot.bar(x='State', y='COVID-19 Deaths', rot=90, ax=ax)
ax.set_ylabel('Number of individuals')
plt.tight_layout()
plt.show()
```

```python
states = ["CA", "TX", "FL", "PA","IL","TN","AL","OK","MN","MS","WA","WV" ,"VT"]
num_deaths = [106981, 103145, 79834, 52337,38315,29865,21195,18245,15045,14641,14638,8218,914]

xs = range(len(states))
plt.bar(xs, num_deaths, color = 'green', width = 0.5)
plt.xticks(xs, states, size = 'small')

plt.xlabel("State")
plt.ylabel("COVID-19 Deaths")
plt.title("Covid-19 Deaths by State")
plt.show()
```

```python
states = ["CA", "TX", "FL", "PA","IL","TN","AL","OK","MN","MS","WA","WV" ,"VT"]
num_pop = [39538245, 29145428, 21538226, 13002689,12812545,6910786,5024356,3959346,5706504,2961288,7705247,1793755,643085 ]

xs = range(len(states))
plt.bar(xs, num_pop, color = 'teal', width = 0.5)
plt.xticks(xs, states, size = 'small')

plt.xlabel("State")
plt.ylabel("Population Size")
plt.title("Population by State")
plt.show()
```

```python
states = ["CA", "TX", "FL", "PA","IL","TN","AL","OK","MN","MS","WA","WV" ,"VT"]
num_deaths = [106981, 103145, 79834, 52337,38315,29865,21195,18245,15045,14641,14638,8218,914]
num_pop = [39538245, 29145428, 21538226, 13002689,12812545,6910786,5024356,3959346,5706504,2961288,7705247,1793755,643085 ]

#divides covid deaths by population to get the death rate in the list
res = [i / j for i, j in zip(num_deaths, num_pop)]

#creates a dictionary with the states and corresponding death rate
statesDeathRates = {}
for key in states:
    for value in res:
        statesDeathRates[key] = value
        res.remove(value)
        break
#sorts the values in the dictionary
sorted_statesDeathRates = {}
sorted_values = sorted(statesDeathRates.values(),  reverse=True)

#sorts the dictionary based on the values
for i in sorted_values:
    for k in statesDeathRates.keys():
        if statesDeathRates[k] == i:
            sorted_statesDeathRates[k] = statesDeathRates[k]

print(sorted_statesDeathRates)

xs = range(len(states))
plt.bar(xs, sorted_statesDeathRates.values(), color = 'blue', width = 0.5)
plt.xticks(xs, sorted_statesDeathRates.keys(), size = 'small')

plt.xlabel("State")
plt.ylabel("COVID-19 Death Rate")
plt.title("COVID-19 Death Rate by State")
plt.show()
```

```python
states = ["CA", "TX", "FL", "PA","IL","TN","AL","OK","MN","MS","WA","WV" ,"VT"]
res = [11.5,13.4,12.4,10.9,11,13.6,14.9,14.3,8.3,18.7,9.5,15.8,9.4]

#creates a dictionary with the states and corresponding poverty rate
statesPovertyRates = {}
for key in states:
    for value in res:
        statesPovertyRates[key] = value
        res.remove(value)
        break
print(statesPovertyRates)

#sorts the values in the dictionary
sorted_statesPovertyRates = {}
sorted_values = sorted(statesPovertyRates.values(),  reverse=True)

#sorts the dictionary based on the values
for i in sorted_values:
    for k in statesPovertyRates.keys():
        if statesPovertyRates[k] == i:
            sorted_statesPovertyRates[k] = statesPovertyRates[k]

xs = range(len(states))
plt.bar(xs, sorted_statesPovertyRates.values(), color = 'red', width = 0.5)
plt.xticks(xs, sorted_statesPovertyRates.keys(), size = 'small')

plt.xlabel("State")
plt.ylabel("Poverty Rate")
plt.title("Poverty Rate by State")
plt.show()
```

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('Covid.csv')
df.drop(columns=['Data As Of', 'Start Date','Condition Group', 'End Date', 'Group','ICD10_codes','Number of Mentions','Flag'], inplace = True)
CovidDeathPerMonth= list(df.iloc[71260:71299, 5])

#creates list for month, year
Month= list(df.iloc[71260:71299, 1])
Year= list(df.iloc[71260:71299, 0])
Xaxis = [str(int(j))+'-'+str(int(i)) for i, j in zip(Month, Year)]

#creates a datetime
date_time = pd.to_datetime(Xaxis)

DF = pd.DataFrame()
DF['value'] = CovidDeathPerMonth
DF = DF.set_index(date_time)
plt.title("COVID-19 Deaths from 1/2020-3/2023")
plt.xlabel('Year/Month') 
plt.ylabel('COVID-19 Deaths') 
plt.plot(DF)
plt.gcf().autofmt_xdate()
plt.show()
```
