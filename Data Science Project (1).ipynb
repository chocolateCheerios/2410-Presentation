```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Covid.csv')
df.drop(columns=['Data As Of', 'Start Date', 'End Date', 'Group', 'Year', 'Month', 'Age Group', 'ICD10_codes', 'Number of Mentions', 'Flag'], inplace=True)

state_counts = df.groupby('State')['COVID-19 Deaths'].sum().reset_index()
state_counts = state_counts[state_counts['State'] != 'United States']
state_counts = state_counts[state_counts['State'] != 'New York City']  # Omit 'New York City'
state_counts = state_counts[state_counts['State'] != 'New York']
state_counts = state_counts.sort_values('COVID-19 Deaths', ascending=False)

# Split state_counts DataFrame into two halves
num_states = len(state_counts)
first_half = state_counts.iloc[:num_states//2, :]
second_half = state_counts.iloc[num_states//2:, :]

# Create first bar chart with first_half DataFrame
plt.figure(figsize=(12, 6))
ax1 = first_half.plot.bar(x='State', y='COVID-19 Deaths', rot=90)
ax1.set_title('COVID-19 Deaths by State (First Half)')
ax1.set_ylabel('Number of individuals')

ax1.set_ylim([2.5, ax1.get_ylim()[1]])  # Set lower limit to 2.5
plt.show()

# Close the first figure before creating the second one
plt.close()

# Create second bar chart with second_half DataFrame
plt.figure(figsize=(12, 6))
ax2 = second_half.plot.bar(x='State', y='COVID-19 Deaths', rot=90)
ax2.set_title('COVID-19 Deaths by State (Second Half)')
ax2.set_ylabel('Number of individuals')
ax2.set_ylim([2.5, ax2.get_ylim()[1]])  # Set lower limit to 2.5
plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Covid.csv')
df.drop(columns=['Data As Of', 'Start Date', 'End Date', 'Group', 'Year', 'Month', 'Age Group', 'ICD10_codes', 'Number of Mentions', 'Flag'], inplace=True)

state_counts = df.groupby('State')['COVID-19 Deaths'].sum().reset_index()
state_counts = state_counts[state_counts['State'] != 'United States']
state_counts = state_counts[state_counts['State'] != 'New York City']
state_counts = state_counts[state_counts['State'] != 'New York']
state_counts = state_counts.sort_values('COVID-19 Deaths', ascending=False)

fig, ax = plt.subplots(figsize=(20, 10))
state_counts.plot.bar(x='State', y='COVID-19 Deaths', rot=90, ax=ax)
ax.set_ylabel('Number of individuals')
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('Covid.csv')
df.drop(columns=['Data As Of', 'Start Date', 'End Date', 'Group', 'Year', 'Month', 'Age Group', 'ICD10_codes', 'Number of Mentions', 'Flag'], inplace=True)

# Aggregate data by state
state_counts = df.groupby('State')['COVID-19 Deaths'].sum().reset_index()
state_counts = state_counts[state_counts['State'] != 'United States']
state_counts = state_counts.sort_values('COVID-19 Deaths', ascending=False)

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(state_counts['COVID-19 Deaths'], labels=state_counts['State'], autopct='%1.1f%%')
plt.title('COVID-19 Deaths by State')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('Covid.csv')

# Drop unnecessary columns
df.drop(columns=['Data As Of', 'Start Date', 'End Date', 'Group', 'Year', 'Month', 'ICD10_codes', 'Number of Mentions', 'Flag'], inplace=True)

# Group the data by Age Group and sum the COVID-19 Deaths
age_counts = df.groupby('Age Group')['COVID-19 Deaths'].sum().reset_index()

# Set the order of the Age Group column
age_counts = age_counts.reindex([7, 6, 5, 4, 3, 2, 1, 0, 8])

# Plot the bar chart
plt.figure(figsize=(10, 6))
ax = age_counts.plot.bar(x='Age Group', y='COVID-19 Deaths', rot=0)
ax.set_xlabel('Age Group')
ax.set_ylabel('Number of COVID-19 Deaths')
ax.set_title('COVID-19 Deaths by Age Group')
plt.show()

```
