```python
states = ["CA", "TX", "FL", "PA","IL","TN","AL","OK","MN","MS","WA","WV" ,"VT"]
num_deaths = [106981, 103145, 79834, 52337,38315,29865,21195,18245,15045,14641,14638,8218,914]
num_pop = [39538245, 29145428, 21538226, 13002689,12812545,6910786,5024356,3959346,5706504,2961288,7705247,1793755,643085 ]

#divides covid deaths by population to get the death rate in the list
res = [i / j for i, j in zip(num_deaths, num_pop)]

print(res)

statesDeathRates = {}
for key in states:
    for value in res:
        statesDeathRates[key] = value
        res.remove(value)
        break
print(statesDeathRates)

sorted_statesDeathRates = {}
sorted_values = sorted(statesDeathRates.values(),  reverse=True)

for i in sorted_values:
    for k in statesDeathRates.keys():
        if statesDeathRates[k] == i:
            sorted_statesDeathRates[k] = statesDeathRates[k]

print(sorted_statesDeathRates)

xs = range(len(states))
plt.bar(xs, sorted_statesDeathRates.values(), color = 'blue', width = 0.5)
plt.xticks(xs, sorted_statesDeathRates.keys(), size = 'small')

plt.xlabel("State")
plt.ylabel("COVID-19 Death Rate")
plt.title("COVID-19 Death Rate by State")
plt.show()
```
